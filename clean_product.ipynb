```python
# df.isnull().sum()
def extract_max_update_ram(ram_str):
    if ram_str and isinstance(ram_str, str):
        match = re.search(r'\d+', ram_str)
        if match:
            return int(match.group())
    return None 
df['max_ram_upgrade'] = df['max_ram_upgrade'].apply(extract_max_update_ram)

df['max_ram_upgrade'].value_counts()
```

```python
def extract_ram(max_ram):
    match = re.search(r'\d+', max_ram)
    return int(match.group())
df['ram_info'] = df['ram_info'].apply(extract_ram)
df['ram_info'].value_counts()
```

### Extract Price:

```python
def extract_price(price_str):
    match = re.findall(r'\d+', price_str)
    return int(''.join(match))
df['price'] = df['price'].apply(extract_price)
df['price'].head(10)
```

### clean refresh_rate

```python
df['refresh_rate'].value_counts()
```

```python
def extract_refresh_rate(x):
    x = str(x).lower()
    
    if x == "hãng không công bố" :
        return None
    nums = re.findall(r'\d+', x)
    nums = list(map(int, nums))
    return max(nums)
df['refresh_rate'] = df['refresh_rate'].apply(extract_refresh_rate)
df['refresh_rate'].value_counts().sum()
```

### clean release_year

```python
def extract_release_year(year_str):
    if pd.isna(year_str):
        return None
    year = str(year_str)
    return int(year)
df['release_year'] = df['release_year'].apply(extract_release_year)
```

### material

```python
def clean_material(material_str):
    material_str = material_str.lower()
    if "nhựa" in material_str:
        return "plastic"
    elif "kim loại" in material_str or "nhôm" in material_str or "magie" in material_str:
        return "metal"
    elif "carbon" in material_str:
        return "carbon"
    else:
        return "other"

df['material'] = df['material'].apply(clean_material)
```

### CPU name

In this step, I want to extract cpu name from name products by regex

```python
df['name_product'].sample(n= 50)
```

```python
pattern = r'\b(i\d\s*\d+[A-Z]*|Core\s*\d+\s*\d+[A-Z]*|Ultra\s*\d+\s*\d+[A-Z]*|R\d\s*(?:AI\s*)?\d+[A-Z]*)\b'
def extract_cpu(name_product):
    match = re.search(pattern, name_product)
    return match.group() if match else None

df['cpu_name'] = df['name_product'].apply(extract_cpu)
```

```python
df["cpu_name"].value_counts()
```

check products with null cpu name

```python
df['cpu_name'].isna().sum()
```

```python
def extract_cpu_fix(name):
    patterns = [
    r'X1[PE]?\s\d+\s\d+',        # Snapdragon X1, X1P
    r'R\d\s*AI\s*\d+',          # Ryzen AI HX series
    r'RYZEN\s*AI.*?\d+',        # RYZEN AI Special
    r'N\d+',                    # Intel N-series
    r'U\d+\s*\d+[A-Z]*',        # another ultra series
    r'UX?9\s*\d+H'              # # Ultra 9 followed by MSI
]
    for p in patterns:
        match = re.search(p, name, re.IGNORECASE)
        if match:
            return match.group()
    return None

df['cpu_name'] = df['cpu_name'].fillna(
    df['name_product'].apply(extract_cpu_fix)
)
```

### clean physical_dimensions

```python
# df['physical_dimensions'].sample(n= 50)
# In this stage, I will seperate it into 3 columns in order to filter easier
pattern_lengths_mn = r"[Dd]ài\s*(\d+\.?\d*)"
pattern_width_mn = r"[Rr]ộng\s*(\d+\.?\d*)"
pattern_thickness_mn = r"[Dd]ày\s*(\d+\.?\d*)"
pattern_weight_kg = r"(\d+\.?\d*) kg"
df['length_mm'] = df['physical_dimensions'].str.extract(pattern_lengths_mn, expand=False)
df['width_mm']  = df['physical_dimensions'].str.extract(pattern_width_mn, expand=False)
df['depth_mm']  = df['physical_dimensions'].str.extract(pattern_thickness_mn, expand=False)
df['weight_kg'] = df['physical_dimensions'].str.extract(pattern_weight_kg, expand=False)
cols = ['length_mm', 'width_mm', 'depth_mm', 'weight_kg']


for col in cols:
    df[col] = df[col].astype(float)
    
```

## GPU clean

```python
df["gpu"].value_counts()
```

```python

def extract_gpu_name(text):
    if not isinstance(text, str):
        return None

    text = re.sub(r"\(.*?\)", "", text)
    # base on some sign -, ,- to normalize
    text = text.replace("-,", "-").replace(", -", "-")
    # after this steps: normalize will into Card tích hợc-AMD Radeon 820M Graphics
    #or Card rời-NVIDIA GeForce RTX 5060, 8 GB
    if "-" in text:
        #ex: Card tích hợc-AMD Radeon 820M Graphics
        part = text.split("-")[-1]
    # remove vram ex 8GB,...
    part = re.sub(r"\d+\s*GB.*", "", part)

    # clean the rest of noise (remnove ,)
    name = part.strip(" ,-")
    return name if name else None
```

```python
df['gpu_type'] = df['gpu'].str.extract(r'(tích hợp|rời)', expand=False)
df['gpu_ram'] = df['gpu'].str.extract(r'(\d+)\s*GB')[0].astype(float)

df["gpu_name"] = df["gpu"].apply(extract_gpu_name)
df["gpu_name"].isna().sum()
df["gpu_name"].sample(n =50 )
```
